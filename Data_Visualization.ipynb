```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
X=iris.data
X[:5]
```

```python
y=iris.target
y[:5]
```

```python
feature_name=iris.feature_names
feature_name
```

```python
target_name=iris.target_names
target_name
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

formatter = FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.figure(figsize=(5, 4))
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap='winter')
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()
```

```python
features = iris.data.T

plt.scatter(features[2], features[3], alpha=0.2, s=100*features[3], c=iris.target, cmap='summer')
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

plt.tight_layout()
plt.show()
```
